# Load Data

```python
import numpy as np

vectors = np.load('outputs/vectors.npy')
```

```python
np.savetxt('outputs/vectors.csv', vectors, delimiter=',')
```

# 2 Dimensional

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, svd_solver='full')
pca_vectors = pca.fit_transform(vectors)
```

```python
import matplotlib.pyplot as plt

plt.scatter(pca_vectors[:, 0], pca_vectors[:, 1], s=0.05)
plt.show()
```

# 3 Dimensional

```python
from sklearn.decomposition import PCA

pca3d = PCA(n_components=3, svd_solver='full')
pca3d_vectors = pca3d.fit_transform(vectors)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(pca3d_vectors[:, 0], pca3d_vectors[:, 1], pca3d_vectors[:, 2], s=0.05)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
